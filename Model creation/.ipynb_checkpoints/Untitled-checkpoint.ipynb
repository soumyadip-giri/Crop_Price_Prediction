```python
# crop_price_ml_comparison.py

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# ===============================
# 1. Load Dataset
# ===============================
df = pd.read_csv("Datasets/crop_price_dataset.csv")
df["Date"] = pd.to_datetime(df["Date"])

# ===============================
# 2. Feature Engineering
# ===============================
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day

# Encode Crop
le_crop = LabelEncoder()
df["CropEncoded"] = le_crop.fit_transform(df["Crop"])

# Features & Target
features = ["Rainfall(mm)", "Supply(quintals)", "DemandIndex", "Year", "Month", "Day", "CropEncoded"]
X = df[features]
y = df["ModalPrice"]

# Scale features for SVM
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ===============================
# 3. Train Models
# ===============================
models = {
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
    "SVM": SVR(kernel="rbf", C=100, gamma=0.1, epsilon=0.1)
}

results = {}
predictions = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    predictions[name] = y_pred

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    results[name] = {"MAE": mae, "RMSE": rmse, "R2": r2}

# ===============================
# 4. Compare Models & Plot
# ===============================
results_df = pd.DataFrame(results).T
print("\nModel Comparison:")
print(results_df)

# Error comparison
plt.figure(figsize=(8,5))
results_df[["MAE","RMSE"]].plot(kind="bar", figsize=(10,6))
plt.title("MAE & RMSE Comparison")
plt.ylabel("Error Value")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# R comparison
plt.figure(figsize=(8,5))
sns.barplot(x=results_df.index, y=results_df["R2"], palette="viridis")
plt.title("R Score Comparison")
plt.ylabel("R Score")
plt.ylim(0,1)
plt.show()

# ===============================
# 5. Save Best Model
# ===============================
best_model_name = results_df["R2"].idxmax()
best_model = models[best_model_name]

export_obj = {
    "model": best_model,
    "scaler": scaler,
    "label_encoder": le_crop,
    "features": features
}

joblib.dump(export_obj, f"{best_model_name.replace(' ','_')}_crop_price_model.pkl")
print(f"\n Best model ({best_model_name}) saved as {best_model_name.replace(' ','_')}_crop_price_model.pkl")
```

```python
import pandas as pd
import joblib

# Load model
model_obj = joblib.load("Random_Forest_crop_price_model.pkl")
model = model_obj["model"]
scaler = model_obj["scaler"]
encoder = model_obj["label_encoder"]
features = model_obj["features"]

# Dynamic input
input_crop = input("Enter crop name (Rice, Potato, Onion, Wheat): ")
input_rainfall = float(input("Enter rainfall in mm: "))
input_supply = float(input("Enter supply in quintals: "))
input_demand = float(input("Enter demand index (0-1): "))
input_month = int(input("Enter month (1-12): "))
input_day = int(input("Enter day (1-31): "))

# Encode crop
crop_encoded = encoder.transform([input_crop])[0]

# Create input dataframe with Year (fixed value)
input_df = pd.DataFrame([{
    "Rainfall(mm)": input_rainfall,
    "Supply(quintals)": input_supply,
    "DemandIndex": input_demand,
    "Year": 2025,          # Keep this so features match
    "Month": input_month,
    "Day": input_day,
    "CropEncoded": crop_encoded
}])

# Scale features
X_new = scaler.transform(input_df[features])

# Predict
predicted_price = model.predict(X_new)[0]
print(f"\nPredicted {input_crop} Price: {predicted_price:.2f}/kg")
```

```python
# ===============================
# 5. Save Best Model
# ===============================
best_model_name = results_df["R2"].idxmax()
best_model = models[best_model_name]

export_obj = {
    "model": best_model,
    "scaler": scaler,
    "label_encoder": le_crop,
    "features": features
}

joblib.dump(export_obj, f"{best_model_name.replace(' ','_')}_crop_price_model.pkl")
print(f"\n Best model ({best_model_name}) saved as {best_model_name.replace(' ','_')}_crop_price_model.pkl")
```
